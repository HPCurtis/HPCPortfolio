```julia
# Load relevant Julia analysis packages.
using CSV, DataFrames, HypothesisTests, Plots, Statistics
```

```julia
url = "https://raw.githubusercontent.com/HPCurtis/HPCPortfolio/main/Data/Turning_Hands_Data_Final.csv"
df = CSV.read(download(url), DataFrame);
first(df, 2)
```

```julia
# Extract the mean_NEo scores and the Rotation column that specifies,
# the condition of the experiment.
red_df = select(df, [:mean_NEO, :Rotation]);
first(red_df, 4)
```

```julia
counter = filter(row -> row[:Rotation] == "counter", red_df)
clock = filter(row -> row[:Rotation] == "clock", red_df)

# Calculate both a equal and unequal variance t-test and output results
# for evaluation.
print(EqualVarianceTTest(counter[:, :mean_NEO], clock[:, :mean_NEO]))
print(UnequalVarianceTTest(counter[:, :mean_NEO], clock[:, :mean_NEO]))
```

```julia
histogram(Any[counter[:, :mean_NEO], clock[:, :mean_NEO]], labels = ["counter" "clock"])
```
